選取作為X的資料 <br/>
Dprice ->股票價格的變化量 <br/>
P1Value ->股票的總值 <br/>
P2Asset ->對手的總資產 <br/>
先選Herd vs. Hedge的data


```python
with open('Herd_Hedge.csv', newline='') as csvfile:
    data = pd.read_csv(csvfile)
X = data.loc[:,['P1Cash','P1Stock','P2Asset','Price','Dprice']]
X = X[1:100]
X = preprocessing.scale(X)
Y = data.loc[:,'P1Decision']
Y = Y[1:100]
Y_dum = pd.get_dummies(Y)
```

```python
np.random.seed(408)
model = Sequential()
model.add(Dense(80, input_dim = 5, activation='relu'))
model.add(Dense(3, activation='sigmoid'))
model.summary()

model.compile(loss='categorical_crossentropy', 
              optimizer='adam', 
              metrics=['accuracy'])

test = model.fit(X,Y_dum, epochs = 500, batch_size = 80, validation_split = 0.2, shuffle = True)
```

```python
plt.plot(test.history['accuracy'])
plt.plot(test.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
plt.plot(test.history['loss'])
plt.plot(test.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```
